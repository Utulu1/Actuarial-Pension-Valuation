```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
ages = [55, 60, 65, 70]

annual_pension = [
    12000,
    15000,
    18000,
    20000
]

pension_data = pd.DataFrame({
    "Retirement_Age": ages,
    "Annual_Pension": annual_pension
})

pension_data
```

```python
discount_rate = 0.04

pension_data["Present_Value"] = (pension_data["Annual_Pension"] / (1 + discount_rate)
                                 
)

pension_data
```

```python
plt.figure(figsize = (8,5))

plt.bar(
    pension_data["Retirement_Age"],
    pension_data["Annual_Pension"]
)

plt.title("Annual Pension by Retirement Age")
plt.xlabel("Retirment Age")
plt.ylabel("Annual Pension (£)")

plt.savefig(
    r"c:\Files fromOneDrive\ActuarialProjects\Actuarial-Pension-Valuation\visualizations\annual_pension_by_age.png"
)
plt.show()
```
